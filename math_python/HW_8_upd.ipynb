### Задание 1

В магазин привезли устройства с 3-х разных предприятий. 

Соотношение устройств следующее: 20% - продукция первого предприятия, 30% - продукция второго предприятия, 50% - продукция третьего предприятия; далее, 10% продукции первого предприятия высшего сорта, на втором предприятии - 5% и на третьем - 20% продукции высшего сорта. 

Найти вероятность того, что случайно купленная новая продукция окажется высшего сорта.

Для решения применим формулу полной вероятности:
P(высший_сорт_общее) = П1*P(П1_ВС) + П2*P(П2_ВС) + П3*P(П3_ВС) = 0,2*0,1 + 0,3*0,05 + 0,5*0,2 = 0,135 или 13,5% - вероятность покупки продукции высшего сорта в магазине

### Задание 2


Пусть брошены 3 уравновешенные монеты.

Рассмотрим 3 события:

A1 - монеты 1 и 2 упали одной и той же стороной;

A2 - монеты 2 и 3 упали одной и той же стороной;

A3 - монеты 1 и 3 упали одной и той же стороной.

Покажите, почему эти 3 события (A1, A2, A3) являются попарно независимыми, но не являются независимыми в совокупности.

Всего возможных исходов выпадения монеты - 8 (2x2x2)

Вероятности событий: P(A1),P(A2),P(A3) = 4/8 = 1/2

Проверим попарную независимость:
1) A1 и А2 - P(A1) x P(A2) = 1/2 x 1/2 = 1/4 - равны, то есть независимы
2) A1 и А3 - P(A1) x P(A3) = 1/2 x 1/2 = 1/4 - равны, то есть независимы
3) A2 и А3 - P(A2) x P(A3) = 1/2 x 1/2 = 1/4 - равны, то есть независимы

Проверка независимости в совокупности:
P(A1) x P(A2) x P(A3) = 1/2 x 1/2 x 1/2 = 1/8 -> 1/4 не равна 1/8, то есть события не являются независимыми в совокупности

События A1,A2,A3 попарно независимы, но в совокупности зависимы.

### Задание 3


Изучите 3 распределения из библиотеки scipy: [Нормальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), [Экспоненциальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html), [Стьюдента](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.t.html). Используя функцию плотности распредления (`pdf`), постройте пять графиков плотностей для каждого распределения при разных параметрах. Запишите ваши наблюдения: как и на что влияют параметры у каждого распределения. 

Дополнительная информация [тут](https://pythonguides.com/scipy-stats/).

```
# Нормальное распределение

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, t

x = np.linspace(-5, 5, 1000)
params = [(0, 1),(0, 0.5),(0, 2),(-1, 1),(1, 0.7)]

plt.figure(figsize=(12, 6))
for i, j in params:
    plt.plot(x, norm.pdf(x, i, j), label=f'μ={i}, σ={j}')

plt.title('Нормальное распределение при различных параметрах')
plt.xlabel('Значение')
plt.ylabel('Плотность вероятности')
plt.legend()
plt.grid()
plt.show()
```

```
# Экспоненциальное распределение

x = np.linspace(0, 5, 1000)
params = [0.5, 1, 1.5, 2, 2.5]

plt.figure(figsize=(12, 6))
for p in params:
    plt.plot(x, expon.pdf(x, scale=p), label=f'λ={1/p:.1f}')

plt.title('Экспоненциальное распределение при различных параметрах')
plt.xlabel('Значение')
plt.ylabel('Плотность вероятности')
plt.legend()
plt.grid()
plt.show()
```

```
# Распределение Стьюдента

x = np.linspace(-5, 5, 1000)

params = [1, 2, 5, 10, 30]

plt.figure(figsize=(12, 6))
for p in params:
    plt.plot(x, t.pdf(x, p), label=f'df={p}')

plt.title('t-распределение при разных степенях свободы')
plt.xlabel('Значение')
plt.ylabel('Плотность вероятности')
plt.legend()
plt.grid()
plt.show()
```
