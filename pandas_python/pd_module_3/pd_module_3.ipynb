```python
import pandas as pd
```

```python
# task_1
log = pd.read_csv("visit_log.csv", sep=";")


def local_traffic_source(country):
    inner_filter = log["traffic_source"].isin(["paid", "email"])
    if country == "Russia":
        return (log["region"] == country) & inner_filter
    else:
        return ~(log["region"] == country) & inner_filter


log.loc[log["traffic_source"].isin(["yandex", "google"]), "source_type"] = "organic"

log.loc[local_traffic_source("Russia"), "source_type"] = "ad"

log.loc[local_traffic_source("not Russia"), "source_type"] = "other"

log["source_type"] = log["source_type"].fillna(log["traffic_source"])
```

```python
# task_3
ratings_df = pd.read_csv("ratings.csv")

user_counts = ratings_df.groupby("userId")["userId"].transform("count")
ratings_df = ratings_df[user_counts > 100]

max_date = pd.to_datetime(
    ratings_df.groupby("userId")["timestamp"].max().rename("max_date"), unit="s"
)
min_date = pd.to_datetime(
    ratings_df.groupby("userId")["timestamp"].min().rename("min_date"), unit="s"
)

users_lifetime_df = pd.concat([max_date, min_date], axis=1)
users_lifetime_df["user_lifetime"] = (
    users_lifetime_df["max_date"] - users_lifetime_df["min_date"]
)
users_lifetime_df
```
