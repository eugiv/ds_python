### Задание 1

Дан объект в $2D$ пространстве

````python
import numpy as np
import matplotlib.pyplot as plt

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

def plot(A):
    x = A[:,0]
    y = A[:,1]
    plt.plot(x, y)
    plt.ylim([-200, 200])
    plt.xlim([-200, 200])

plot(A)    
````

При помощи линейных отображений:
- Уменьшить объект в два раза
- Повернуть на 130 градусов
- Отразить объект относительно прямой y=x (поможет матрица `[[0, 1, 0], [1, 0, 0], [0, 0, 1]]`)


````python
# 1
A_reduced = A.copy().astype(np.float64)
A_reduced[:, :2] *= 0.5

plot(A_reduced)
````

````python
# 2
theta = np.radians(130)
cos_theta, sin_theta = np.cos(theta), np.sin(theta)

rotation_matrix = np.array([
    [cos_theta, -sin_theta, 0],
    [sin_theta, cos_theta, 0],
])

A_rotated = A_reduced @ rotation_matrix.T

plot(A_rotated)
````

````python
# 3
reflect_mat = np.array([[0, 1, 0], [1, 0, 0], [0, 0, 1]])
A_reflected = A @ reflect_mat

plot(A_reflected)
````

### Задание 2

Найдите спектральное разложение матрицы:
```
m = np.array([[1, 2],
              [2, 3]])
print(m)
```



````python
m = np.array([[1, 2],
              [2, 3]])

eigenval, eigenvec = np.linalg.eig(m)

diag_mat = np.diag(eigenval)
reconstructed_mat = eigenvec @ diag_mat @ eigenvec.T

print('Matrix spectral decomposition:\n')
print('Matrix own values: \n', eigenval)
print('Matrix own vectors: \n', eigenvec)
print('Diagonal matrix: \n', diag_mat)
print('Reconstructed matrix: \n', reconstructed_mat)
````
