```python
import pandas as pd

movies = pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings.csv")
```

```python
movies_ratings = pd.merge(ratings, movies, on="movieId", how="left")
```

```python
# task_1
most_rated_movie = movies_ratings[["title", "rating"]].value_counts().idxmax()[0]
print(f"the most rated movie is: {most_rated_movie}")
```

```python
# task_2
power = pd.read_csv("power.csv")
baltic_states_power = power[
    power["country"].str.contains("lithuan|latv|estonia|eesti", case=False)
]
```

```python
baltic_states_power = baltic_states_power[
    (baltic_states_power["quantity"] >= 0)
    & (baltic_states_power["category"].isin([1, 12, 21]))
    & (baltic_states_power["year"].between(2005, 2010))
]
baltic_states_power_ttl_cmpsmtn = baltic_states_power["quantity"].sum()
print(
    f"Total compsumption of power of Baltic states is: {baltic_states_power_ttl_cmpsmtn} MegagiperWt"
)
```
