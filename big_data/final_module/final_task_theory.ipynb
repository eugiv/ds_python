Придумать Data-проект, который должен улучшить показатели Вашего бизнеса и расписать его по Crisp-DM

Data-проект: Прогноз рекомендации:
Цель: Увеличить просмотры и снизить отток.

1. Business Understanding:
Увеличить просмотры на 15%, снизить отток на 10%.

2. Data Understanding:
Логи просмотров, профили пользователей, рейтинги.

3. Data Preparation:
Чистка (удаление кликов <10 сек), добавление жанров/актеров.

4. Modeling:
Рекомендации, классификации
Прогноз спроса

5. Evaluation:
A/B-тест новых рекомендаций против старых.

6. Deployment на прод

Описать требуемые роли в команде по работе с данными на этапах 4 и 5

1. Data Engineer

Настраивает ETL-пайплайны для загрузки данных в хранилище.
Автоматизирует проверки качества.
Обеспечивает обновление данных для рекомендательной системы.

За что отвечает:
Чтобы данные были доступны, свежие и чистые.

2. Data Analyst

Проверяет качество данных (дубликаты, аномалии, выбросы).
Анализирует, какие фильмы чаще бросают/досматривают.
Готовит отчеты для Product Analyst.

За что отвечает:
Чтобы данные были надежными и помогали принимать решения.

3. ML Engineer

Разрабатывает модель рекомендаций и прогноза оттока.
Интегрирует модель в пайплайн данных.

За что отвечает:
Чтобы рекомендации работали в реальном времени.

4. Product Analyst

Определяет ключевые метрики (просмотры, отток).
Анализирует A/B-тест новых рекомендаций.

За что отвечает:
Чтобы проект увеличил прибыль.

5. DevOps Engineer

Развертывает хранилище и ML-модели в продакшене.
Настраивает мониторинг (чтобы всё не упало).

За что отвечает:
Чтобы система работала без сбоев.
