```
def f(x, a, b):
    return np.e**a * np.sin(b * x) + x


a_vals = np.linspace(0, 10, 100)
b_vals = np.linspace(0, 10, 100)

best_error = float("inf")
optimal_params = None

for a in a_vals:
    for b in b_vals:
        try:
            popt, pcov = curve_fit(f, x, y, p0=[a, b])
            error = np.sum((y - f(x, *popt)) ** 2)

            if error < best_error:
                best_error = error
                optimal_params = popt
        except:
            continue

print(optimal_params[0])
print(optimal_params[1])
```
