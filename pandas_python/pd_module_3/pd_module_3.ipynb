```python
import pandas as pd

log = pd.read_csv('visit_log.csv', sep=';')
```

```python
#task_1

def local_traffic_source(country):
    inner_filter = (log["traffic_source"].isin(["paid", "email"]))
    if country == "Russia":
        return (log["region"] == country) & inner_filter
    else:
        return ~(log["region"] == country) & inner_filter

log.loc[log["traffic_source"].isin(["yandex", "google"]), "source_type"] = "organic"

log.loc[local_traffic_source("Russia"), "source_type"] = "ad"

log.loc[local_traffic_source("not Russia"), "source_type"] = "other"

log["source_type"] = log["source_type"].fillna(log["traffic_source"])
log.tail(20)
```
